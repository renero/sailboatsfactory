## Orthogonalization

I've plot the results from a grid search over the hyperparameters space of `timesteps` and `batch_size`, to find that when using Timesteps = 6, and a Batch Size of 8, the trend error (number of trend missclassifications) is minimum (only 1 error).

The number of epochs used is only 10 to minimize computation time. 
The dataset used is the one with all 2017 up to Dec 20th, with hourly values.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
 
# Get the data (csv file is hosted on the web)
url = '/Users/renero/Documents/SideProjects/sailboatsfactory/notebooks/experimental/output_4956.csv'
df = pd.read_csv(url, delimiter=';')
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.gca(projection='3d')
# Axis labels
ax.set_xlabel('Timesteps')
ax.set_ylabel('Batch Size')
ax.set_zlabel('Trend Error')
# Plot3D
surf = ax.plot_trisurf(df['ts'], df['bs'], df['te'], cmap=plt.cm.viridis, linewidth=0.2)
# rotate the axes and update
ax.view_init(30, 45)
plt.show()
```
